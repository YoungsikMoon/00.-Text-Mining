## 정규화 어간 표제어

정규화 : 형태는 다르지만 같은 의미를 가진 단어를 하나로 통일시켜주는 작업

중복된 단어를 하나로 통합하여 데이터를 줄임.

Apple, apple -> apple (소문자로 통일 시켜줌)

갔다, 갔었다 -> 가다

가다, 간다, 갔다, 갔었다, 갈 것이다 -> 가

표제어(Lemmatization) : 단어의 기본형

goes -> go

## 인코딩

벡터화(Vectorize) : 텍스트를 컴퓨터가 이해할 수 있도록 벡터로 만들어주는 것, 텍스트 입력을 수치 표현으로 변환하는 것.

벡터화에는 다양한 방식이 존재하는데 그중 기본적으로 아래 것들이 있다.

BoW(Bag-of-Words), DTM(Document-Term Matrix) ,TF-IDF(Term Frequency-Inverse Document Frequency), Word Embedding

벡터화를 알기 위해선 원-핫 인코딩, 희소표현 이란 개념을 알고가면 좋다.

---


원-핫 인코딩 : 사과 배 수박 가 있다면.

사과 [ 1, 0, 0]
배 [0, 1, 0]
수박 [0, 0, 1]

---

이렇게 되면 되면 적은 단어나 문장을 원-핫 인코딩 하기엔 좋지만.. 문장이 길어지거나 단어가 많아진걸 원-핫 인코딩 하면 0이 무수히 많아진다.

단어 10,000개 가 있다고 한다면..

사과 [0,0,0,0,0,0, ....0,1]
이런걸 '희소'라고 한다.

이렇게 표현하면 데이터관리가 힘들어진다. 이 걸 희소표현(Sparse Matrix) 문제라고 한다.

---




BoW : 단어의 등장 순서는 고려하지 않고, 단어들의 출현 빈도에만 집중하는 텍스트 데이터의 수치화 표현 방법.

---

DTM : 다수의 문서에서 등장하는 각 단어들의 빈도를 행렬로 표현한 것.

---

TF-IDF : 전체 코퍼스와 비교하여 문서에 있는 각 단어의 중요치 계산.

문서 내 단어의 빈도수가 높더라도 다른 문서에서도 많이 등장하는 단어라면 TF-IDF 값은 낮게 나옴.

문서 내 단어의 빈도수가 높은 단어들 가운데, 다른 문서에서는 잘 등장하지 않는 단어의 TF-IDF 값이 높게 나옴.

---

Word Embedding : 희소표현 문제를 개선하여 나온 것.

대표적으로 word2vec, GloVe 등이 있다.

희소표현 대신 밀집표현(Dense Matrix)을 하며. 값은 실수가 된다. 단어의 관계 학습이 필요한 방식이다.

## N-Gram

N-gram : n개의 연속된 단어를 새로운 하나의 단어로 정의

2개의 단어를 하나로 묶으면 2-gram (=bi-gram)

3개의 단어를 하나로 묶으면 3-gram (=tri-gram)

그 이상의 단어를 하나로 묶으면 :벡터 크기가 커지고 희소문제가 생길 수 있어서 2~3 정도만 사용한다.

N-gram 목적 : 단어들의 순서를 고려할 수 있게 하기 위한 방법이다. 완벽하진 않지만, 부분적으로 단어의 순서 정보를 반영하여 분석할 수 있다.

## spaCy 불용어 처리 맛보기

실습에 사용할 패키지 : spaCy

사용한 이유 : 패키지 spaCy는 불용어 사전이 미리 만들어져 있으니 한번 연습삼아 써보기

```python
#우선 spacy가 제공하는 불용어 사전을 불러오자
spacy_stop_words = spacy.lang.en.stop_words.STOP_WORDS

# 불러온 불용어 사전에서 어떤것들이 있는지 5개만 확인해보자
temp_stop_words = [] #임시 불용어 사전
for i, stop_word in enumerate(spacy_stop_words): #enumerate는 인덱스를 부여하는 것임 테스트 편의상 써봄
    if i == 5 :break
    print(stop_word)
    temp_stop_words.append(stop_word) # 지금 확인된 불용어는 regarding nobody below over am 이렇게 5가지를 확인했다.. 분명 더 많겠지만! 이것만 임시 불용어 사전에 담아보자
```

```python
# 임시 불용어는 출력되지 않게 해보자
text = "나는 책을 regarding 하는것이 너무 귀찮아 혹시 nobody 날 도와줄 below 는 없을까? 그럼 이 순간이 over 될 텐뎅" #영어를 몰라서 그냥 막 써본 텍스트임..
filtered_tokens = ' '.join([token.text for token in spacy_en(text) if token.text not in temp_stop_words]) #텍스트를 토큰한 단어들이 불용어 사전에 없으면 토큰을 출력하자.
print(filtered_tokens)
```

## NLTK로 어간 추출(Stemming) 맛보기



다양한 어간 추출기가 존재하지만 앞에서 사용해본 NLTK에서 제공하는 스테머로 학습해보자.

```python
from nltk import PorterStemmer # 단순 규칙에 의해 나눠진 것이므로 사전에 없는 단어가 찍힐 수있음.
stemmer = PorterStemmer() #여러번 테스트 하기위해 스테머  객체 하나 생성해서 stemmer 라는 변수에 담아서 재사용하자.

print(stemmer.stem('working'), stemmer.stem('works'), stemmer.stem('worked'))
print(stemmer.stem('amusing'), stemmer.stem('amuses'), stemmer.stem('amused'))
print(stemmer.stem('happier'), stemmer.stem('happiest'), stemmer.stem('happying'))
print(stemmer.stem('was'), stemmer.stem('loving'))
```

```python
from nltk import LancasterStemmer # 또 다른 스테머 Porter랑 결과 다름 왜 다른지는 nltk 홈페이지를 참고하자
stemmer = LancasterStemmer()
print(stemmer.stem('working'), stemmer.stem('works'), stemmer.stem('worked'))
print(stemmer.stem('amusing'), stemmer.stem('amuses'), stemmer.stem('amused'))
print(stemmer.stem('happier'), stemmer.stem('happiest'), stemmer.stem('happying'))
print(stemmer.stem('was'), stemmer.stem('loving'))
```

```python
from nltk import RegexpStemmer # 사용자가 지정한 정규 표현 방식으로 어간 추출함
stemmer = RegexpStemmer('ing') #ing 처리
print(stemmer.stem('amusing'), stemmer.stem('amuses'), stemmer.stem('amused'))
print(stemmer.stem('happier'), stemmer.stem('happiest'), stemmer.stem('happying'))
print(stemmer.stem('was'), stemmer.stem('loving'), stemmer.stem('sing')) #sing 에서 ing 를 빼니 s만 남음 ㅋ
```

```python
from nltk import SnowballStemmer # 영어 외 13개 국어 stemming
stemmer = SnowballStemmer('spanish') #스페인어
print(stemmer.stem('hola')) #홀라의 어간은 hol 인가?
```

```python
from nltk import SnowballStemmer
stemmer = SnowballStemmer('english')
print(stemmer.stem('fishing'),stemmer.stem('fished'),stemmer.stem('fishs'))
```

## NLTK 표제어 추출(Lemmatization) 맛보기




표제어(Lemma)는 한글로 '기본 사전형 단어' 정도의 의미를 갖는다.

am are is 는 다른 형태를 가졌지만 뿌리는 be 로 같습니다.

* 형태소: 의미를 가진 가장 작은 단위!! 혼동하지 말것. (난 좀 자주 헷갈린다)


```python
from nltk import WordNetLemmatizer
# 표제어 추출에 필요한 데이터 다운로드
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
lemma = WordNetLemmatizer()
# v는 단어가 동사일 때 동사의 기본형을 얻는다.
# a는 단어가 형용사일 때 형용사의 기본형을 얻는다.
print(lemma.lemmatize('amusing'),lemma.lemmatize('amuses'),lemma.lemmatize('amused'))
print(lemma.lemmatize('happier'),lemma.lemmatize('happiest'))
print(lemma.lemmatize('fancier'),lemma.lemmatize('fanciest'))
print(lemma.lemmatize('was'), lemma.lemmatize('love'))
print(lemma.lemmatize('amusing','v'),lemma.lemmatize('amuses','v'),lemma.lemmatize('amused','v'))
print(lemma.lemmatize('happier','a'),lemma.lemmatize('happiest','a'))
print(lemma.lemmatize('fancier','a'),lemma.lemmatize('fanciest','a'))
print(lemma.lemmatize('was', 'v'), lemma.lemmatize('love', 'v'))
print(lemma.lemmatize('dies','v'))
print(lemma.lemmatize('watched', 'v'))
print(lemma.lemmatize('has', 'v'))
```

```python
#정규 표현식을 사용하기 위한 라이브러리 호출
import re
```

```python
# . 은 한 개의 임의의 문자를 나타낸다.
# 예를 들어 정규 표현식이 a.c라고 한다면, a와 c 사이에 어떤 1개의 문자라도 올 수 있다는 뜻이다.'
r = re.compile("a.c")
print(r.search("kkk")) #매치 안됨.
print(r.search("abc")) #매치 됨.
```

```python
# ? 는 ? 앞의 문자가 존재할 수도 있고, 존재하지 않을 수도 있는 경우를 나타낸다.
# 예를 들어 정규 표현식이 ab?C라고 한다면, b는 있다고 취급할 수도 있고 없다고 취급할 수도 있다. 즉, abc와 ac 모두 매치 가능
r = re.compile("ab?c")
print(r.search("abbc")) #매치 안됨.
print(r.search("abc")) #매치 됨.
print(r.search("ac")) #매치 됨.
```

```python
# * 는 바로 앞의 문자가 0개 이상일 경우를 나타낸다.
# 앞의 문자는 존재하지 않을 수도 있으며, 또는 여러 개일 수도 있다.
# 예를 들어 정규 표현식이 abc라고 한다면 ac, abc, abbc, abbbc 등과 매치할 수 있다.
r = re.compile("ab*c")
print(r.search("a"))
print(r.search("ac"))
print(r.search("abc"))
print(r.search("abbc"))
```

```python
# + 는 *와 유사하다.
# 다른 점은 앞의 문자가 모두 있어야 한다.
# 예를 들어 ab+c라고 하면, ac는 매치되지 않는다
r = re.compile('ab+c')
print(r.search("ac"))# 아무것도 출력되지 않는다.
print(r.search("abc"))
```

```python
# ^ 는 시작되는 글자를 지정한다.
# 예를 들어 ^a라면 a로 시작되는 문자열만 찾는다
r = re.compile('^a')
print(r.search('bbc')) # 아무것도 출력되지 않음
print(r.search('ab'))
```

```python
# $는 마지막 글자를 지정한다.
# 예를 들어 $a라면 a로 끝나는 문자열을 찾는다
r = re.compile('a$')
print(r.search('aasdasdasd')) #아무것도 출력되지 않음
print(r.search('aasdasdasda'))
```

```python
# 문자{숫자1, 숫자2}
# 문자에 해당 기호를 붙이면, 해당 문자를 숫자1 이상 숫자2 이하만큼 반복한다.
# 예를 들어 정규 표현식이 ab{2,8}c라면 a와 c사이에 b가 존재하면서 b는 2개 이상 8개 이하인 문자열에 대해서 매치한다.
r = re.compile("ab{2,8}c")
print(r.search('ac'))# 아무런 결과도 출력되지 않음
print(r.search("abc"))# 아무런 결과도 출력되지 않음
print(r.search("abbc"))
print(r.search("abbbbc"))
print(r.search("abbbbbbbbbbbc")) # 아무런 결과도 출력되지 않음
```

```python
# 문자{숫자1,}
# 문자에 해당 기호를 붙이면, 해당 문자를 숫자 이상만큼 반복한다.
# 예를 들어 정규 표현식이 a{2,}bc라면 뒤에 bc가 붙으면서 a의 갯수가 2개 이상인 경우인 문자열과 매치한다.
# 또한 만약 {0,}을 쓴다면 *와 동일한 의미가 되며, {1,}을 쓴다면 +와 동일한 의미가 된다.
r = re.compile("a{2,}bc")
print(r.search("bc")) # 아무런 결과도 출력되지 않음
print(r.search("aa")) # 아무런 결과도 출력되지 않음
print(r.search("aabc"))
print(r.search("aaaaabc"))
```

```python
# [ ]
# []안에 문자들을 넣으면 그 문자들 중 한 개의 문자와 매치라는 의미를 가진다.
# 예를 들어 정규 표현식이 [abc]라면, a 또는 b 또는 c가 들어가 있는 문자열과 매치된다.
# 범위를 지정하는 것도 가능하다. [a-zA-Z]는 알파벳 전부를 의미하며, [0–9]는 숫자 전부를 의미한다.
r = re.compile("[abc]")
print(r.search('zzz'))# 아무것도 출력되지 않음
print(r.search('a'))
```

[a-c] : [abc]와 같음

[0-5] : [012345]와 같음

[a-zA-Z] : 모든 알파벳

[0-9] : 모든 숫자

```python
# ^[문]
# 앞서 설명한 ^와는 완전히 다른 의미로 쓰인다.
# 여기서는 ^ 기호 뒤에 붙은 문자들을 제외한 모든 문자를 매치하는 역할을 한다.
# 예를 들어 [^abc]라는 정규 표현식이 있다면, a 또는 b 또는 c가 들어간 문자열을 제외한 모든 문자열을 매치한다.

r = re.compile('[^abc]')
print(r.search("a")) # 아무것도 출력되지 않음
print(r.search("ahoho"))
```

```python
# re.match() 와 re.search() 차이 알아보기
# search는 정규표현식 전체에 대해서 문자열이 매치하는지 본다.
# match는 문자열 첫 부분부터 정규표현식과 매치하는지 본다.

r = re.compile("ab.")
print(r.search("kkkabc"))
print(r.match("kkkabc")) # 출력 x
print(r.match("abckkk"))
```

```python
# re.split() 은 입력된 정규 표현식을 기준으로 문자열들을 분리하여 리스트로 리턴한다.
# 토큰화에 유용하게 쓰일 수 있기 때문에 자연어 처리에서 가장 많이 사용되는 정규 표현식 함수 중 하나.
text = "사과 딸기 수박 멜론 바나나"
print(re.split(" ",text))

text = """사과
딸기
수박
멜론
바나나"""
print(re.split("\n", text))

text = "사과+딸기+수박+멜론+바나나"
print(re.split("\+", text))
print(text.split('+'))
```

```python
# re.findall() 은 정규 표현식과 매치되는 모든 문자열들을 리스트로 리턴한다.
# 단, 매치되는 문자열이 없다면 빈 리스트를 리턴한다.

text = """이름 : 김철수
전화번호 : 010 - 1234 - 1234
나이 : 30
성별 : 남"""
print(re.findall("\d+", text))
print(re.findall("\d+", "문자열입니당.")) # 빈 리스트를 리턴했다.
```

```python
# re.sub() 은 정규 표현식 패턴과 일치하는 문자열을 찾아 다른 문자열로 대체할 수 있다.
text="Regular expression : A regular expression, regex or regexp[1] (sometimes called a rational expression)[2][3] is, in theoretical computer science and formal language theory, a sequence of characters that define a search pattern."
print(re.sub('[^a-zA-Z]',' ',text)) #영어가 아닌 것들은 공백으로 교체 됨.
```

## 정규 표현식을 활용한 텍스트 전처리 맛보기

아래 코드를 실행하며 출력을 보고 고민해보자

```python
# 아래 텍스트로 정규표현식 토큰화를 해보겠다.
text = """100 John    PROF
101 James    STUD
102 Mac    STUD"""
```

```python
# \s 는 공백 여러개를 의미한다.
print(re.split('\s+', text))
```

```python
# \d 는 숫자를 의미한다.
print(re.findall('\d+', text))
```

```python
# [A-Z]는 영어 대문자를 의미한다.
print(re.findall('[A-Z]', text))
```

```python
# [^a-zA-Z] 영어가 아닌 것을 의미한다.
print(re.sub('[^a-zA-Z]', ' ', text))
```

```python
num1 = '123-456-7890'
num2 = '123 456 7890'
num3 = '1234567890'
pattern = '\d+[-\s]?\d+[-\s]?\d+' # 사람마다 번호 기입 방식이 다른데 모두 처리 가능하도록 정규화 함.
numbers1 = re.search(pattern, num1)
numbers2 = re.search(pattern, num2)
numbers3 = re.search(pattern, num3)
print(numbers1.group())
print(numbers2.group())
print(numbers3.group())
```

```python
# "| : OR(논리합)을 의미
# '*'는 그 앞의 글자가 0회 이상 나타남을 의미(즉, 그 글자가 한번도 안 나와도 되고, 여러 번 나와도 됨.)
# 아래 정규식은 'hello', 'hey'로 시작
# 그 다음에 임의의 개수의 빈칸(space)들
# 임의의 개수의 영어 문자들이 오는 인사말이다.
r = "(hi|hello|hey)[ ]*([a-z]*)"
```

```python
# 영어 인사만 받는다 한국말은 걸러진다.
print((re.search(r, "hello everyone? 짜라빠빠")))
print((re.search(r, "who everyone?")))
print((re.search(r, "why everyone?")))
print((re.search(r, "그대는 아름다워 hi everyone?")))
print((re.search(r, "Hey everyone?"))) # Hey 와 hey는 다르다.
```

```python
# 영문 텍스트의 경우에는 정규식을 단순하게 유지하기 위해 이처럼 대소문자 구분을 무시할 때가 많다.
# IGNORECASE : 대소문자 구별 없이 매치를 수행할 때 사용하는 옵션
print(re.match(r, "Hello Rosa", flags = re.IGNORECASE)) # 대문자를 소문자로 전환해서 Hello -> hello로 인식
```

```python
# 아래의 방법을 사용하는 경우, 텍스트의 첫 글자는 직접 소문자로 변경해야 합니다.
text = 'Today is Monday. Tomorrow is Tuesday. Yesterday was Sunday.'
first_big_words = re.findall(r'[\.\?\!]\s+([A-Z]\w+)', text) #첫 글짜가 대문자인 것만 추출
first_small_word = [word.lower() for word in first_big_words] #소문자로 변경
print(first_big_words)
print(first_small_word)
```

## 네이버 영화 리뷰 데이터로 워드클라우드 맛보기2

워드클라우드를 어떻게 더 개선해 볼 것인가?

1. 한글 아닌 데이터 제거 *정규화 활용
2. 문장을 단어로 나누기 *토큰화 활용 (+단어 형태소 추출)
3. 불용어 제거



---



```python
# 다운 받은 데이타 활용 이번엔 5만 데이터로 진행해보겠다, 데이터가 크면 오래걸린다..
import pandas as pd
DATA_PATH = '/content/'
rating_data = pd.read_csv(DATA_PATH+'ratings_test.txt', header = 0, delimiter='\t', quoting=3)
```

```python
rating_data['document'][:5] # 데이터 확인
```

```python
#전처리(한글 아닌 데이터 제거 + 형태소 토큰화, 불용어 제거) 함수 만들기
import re #정규화 패키지
def preprocessing(review, okt, remove_stopwords = False, stop_words =[]):
  # review: 전처리할 텍스트 데이터
  # okt: okt객체를 반복적으로 생성하지 않고 미리 생성 후 인자로 받음
  # remove_stopword: 불용어를 제거할지 여부 선택. 기본값 False
  # stop_words: 불용어 사전은 사용자가 직접 입력, 기본값 빈 리스트

  # 1. 한글 및 공백 제외한 문자 모두 제거
  review_text = re.sub('[^가-힣ㄱ-ㅎㅏ-ㅣ\\s]','',review)

  #2. okt 객체를 활용하여 형태소 단어로 나눔
  word_review = okt.morphs(review_text,stem=True)

  if remove_stopwords:
    #3. 불용어 제거(선택)
    word_review = [token for token in word_review if not token in stop_words]
  return word_review #정제된 코퍼스
```

```python
# 불용어 정의
stop_words = ['은','는','이','가','하','아','것','들','의','있','되','수','보','주','등','한']

!pip install konlpy
from konlpy.tag import Okt
okt = Okt() # NLTK의 Okt 생성
```

```python
clean_review = [] #전처리 된 리뷰 데이터가 담길 리스트
for review in rating_data['document']:# 전처리 된 코퍼스 3분 소요 됐음
  # 리뷰가 문자열인 경우만 전처리 진행
  if type(review) == str:
    clean_review.append(preprocessing(review,okt,remove_stopwords=True,stop_words= stop_words))
  else:
    clean_review.append([]) #str이 아닌 행은 빈칸으로 놔두기
```

```python
clean_review = sum(clean_review,[]) #sum(리스트,[]) 이렇게 하면 리스트 안의 리스트들을 하나로 합칠 수 있다.
```

```python
# 한글 폰트 설정
from wordcloud import WordCloud #워드클라우드 라이브러리
wordcloud=WordCloud(font_path=fontpath).generate(' '.join(clean_review))
```

그럼 겸사겸사 '비지도학습' 이라는 개념도 같이 알고 넘어갈 수 있다.


---




SOYNLP는 비지도 학습기반의 품사 태깅 단어 토큰나이저 이다.

지도/비지도 아주 쉽게 줄여 말하면

지도 학습 : 정답이 있는 학습

비지도 학습 : 정답이 없는 학습

아래를 보면서 비지도를 느껴보자.

----

SOYNLP의 특징 정리

1. 비지도 학습으로 단어를 토큰화 한다.

2. 데이터에 자주 등장하는 단어를 단어로 분석한다. (말이 이상해보이는데 아래 코드를 따라가면 이해가능)

3. 내부적으로 단어 점수 표로 동작한다.

4. 점수는 응집 확률, 브랜칭 엔트로피를 활용한다. (짧고 간단하게 설명하겠다)



---



응집확률 : 특정 문자열이 주어졌을 때 다음 문자가 얼마나 같이 등장하는지를 판단.
* 친구들이 얼마나 끈끈하게 잘 어울려 놀고 있는지를 점수로 표현한다고 상상해보자.

브랜칭 엔트로피 : 특정 문자열에서 다음에 올 수 있는 문자의 '다양성'을 측정하는 방법.
* '사과는 맛있다' 다음에 올 수 있는 단어는 '과일', '음식' 등 다양할 수 있다.
* '맛있는' 다음에 올 수 있는 다양성이 높기 때문에, 이 지점의 브랜칭 엔트로피(불확실성의척도) 값도 높다.

---



왜 비지도학습 방식 토큰화가 생겼을까? 라는 의문이 드는 사람도 있을 것이다.

간단하게 이 부분도 설명해주겠다.

기존 형태소 분석기가 가지는 문제는 무엇인가? 기존 형태소 분석기는 분석기에 등록되지 않은 단어 예를들면 신조어 같은 경우에는 제대로 구분하지 못하는 단점이 있었다.

### 정수 인코딩 맛보기 (Dictionary)




딕셔너리를 활용한 정수 인코딩을 알아보자.

```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
nltk.download('popular')
```

```python
text = "A barber is a person. a barber is good person. a barber is huge person. he Knew A Secret! The Secret He Kept is huge secret. Huge secret. His barber kept his word. a barber kept his word. His barber kept his secret. But keeping and keeping such a huge secret to himself was driving the barber crazy. the barber went up a huge mountain."
text = sent_tokenize(text)# 문장 토큰화
print(text)
```

```python
#정제와 단어 토큰화
vocab = {} # 단어 : 빈도수 딕셔너리
sentences = []
stop_words = set(stopwords.words('english')) #불용어 불러오기
for i in text:
    # 토큰화 된 문장이 있으면 수행 토큰화 된 문장을 i에 넣고 수행,
    # 한번 for문 수행 하고 다 끝나면 다음 토큰화 된 문장이 있으면 또 수행
    sentence = word_tokenize(i) #단어 토큰화를 수행
    result = []
    for word in sentence:
        word = word.lower() # 모든 단어를 소문자화 함
        if word not in stop_words: # 해당 단어가 불용어 인지 체크
            if len(word) > 2: # 2글자 넘는 단어인지 체크
                result.append(word)
                if word not in vocab: #빈도수 사전에 단어가 없으면
                    vocab[word] = 0 #빈도수 사전에 추가후 벨류는 0으로 지정
                vocab[word] += 1 #단어 빈도수 1 증가
    sentences.append(result)
```

```python
print(sentences) #문장 토큰 별로 단어가 어떻게 있는지 확인
```

```python
print(vocab) #단어:빈도수 매핑된 사
```

```python
#빈도수가 높은 순서대로 정렬
vocab_sorted = sorted(vocab.items(), key = lambda x:x[1], reverse = True)
print(vocab_sorted)
```

```python
#높은 빈도수를 가진 단어일수록 낮은 정수 인덱스를 부여
word_to_index = {} #빈도수 높은 단어 순으로 저장
i=0
for (word, frequency) in vocab_sorted :
    if frequency > 1 : #빈도수가 1이하 단어는 제외한다.
        i=i+1
        word_to_index[word] = i
print(word_to_index)
```

```python
vocab_size = 5 #빈도수 랭킹 5위 까지만 사용하려고.
words_frequency = [w for w,c in word_to_index.items() if c >= vocab_size + 1] # 인덱스가 5 초과인 단어 제거
for w in words_frequency:
    del word_to_index[w] # 해당 단어에 대한 인덱스 정보를 삭제
print(word_to_index)
```

여기 까지 진행하면 word_to_index에는 빈도수가 높은 상위 5개의 단어만 저장된다.

 word_to_index(빈도수가 높은 순서로 인덱스:단어 쌍으로 저장된 딕셔너리)를 사용하여 단어 토큰화가 된 상태로 저장된 sentences에 있는 각 단어를 정수로 바꾸는 작업을 하겠다.

예) sentences에서 첫번째 문장은 ['barber', 'person']이었는데, 이 문장에 대해서는 [1, 5]로 인코딩

근데 두번째 문장인 ['barber', 'good', 'person']에는 더 이상 word_to_index에는 존재하지 않는 단어인 'good'이라는 단어가 있다.

이처럼 단어 집합에 존재하지 않는 단어들을 Out-Of-Vocabulary(단어 집합에 없는 단어)의 약자로 'OOV'라고 한다.

word_to_index에 'OOV'란 단어를 새롭게 추가하고, 단어 집합에 없는 단어들은 'OOV'의 인덱스로 인코딩한다.


어려워 보이지만 코드를 추적하다보면 이해가 될 것이다.
'

---



```python
word_to_index['OOV'] = len(word_to_index) + 1 #5위 까지만 했으니 OOV는 6이다.
```

```python
#word_to_index['OOV']에 sentences의 빈도수 적은 모든 단어들을 맵핑되는 정수로 인코딩
encoded = []
for s in sentences:
    temp = []
    for w in s:
        try:
            temp.append(word_to_index[w])
        except KeyError: #word_to_index에 순위가 낮은 단어가 없으니 에러가 발생한다.
            temp.append(word_to_index['OOV']) #단어를 OOV 값을 temp에 넣는다. 즉, 6 이다.
    encoded.append(temp)
print(sentences)
print(encoded)
```

## 통계적 언어 모델 개념 (SLM)

통계적 언어 모델(Statistical Language Model, SLM)

---



통계적 언어 모델은 많은 텍스트 데이터(코퍼스)를 학습하여, 실제 언어에서 단어들이 어떻게 조합되어 나타나는지 확률 분포를 근사한다.

개념 몇 가지를 알아보자.

---



연쇄 법칙  P(x1,x2,x3,⋯,xn)=P(x1)P(x2|x1)P(x3|x1,x2)…P(xn|x1,…,xn−1)

다시 말하지만 난 수식을 봐도 모르는 사람이다..

수학 잘 하는 사람은 수식을 보고 이해하겠지만 나같은 사람은 그냥 감각적으로 느끼고 풀어 해석하는게 빠르다.

나와 함께 감각적으로 이해해볼 사람들은 아래를 읽어라.

기본 아이디어: 연쇄법칙은 복잡한 문제를 '작은 부분'으로 나누어 해결하는 방법.

문장에서 다음 단어를 예측할 때, 각 '단어'가 나올 확률을 조금씩 계산하여 '전체 문장'의 확률을 구하는 방식.

---



단어 예측

예를 들어, "나는 아침에"라는 문장이 있을 때, "밥을"이라는 단어가 다음에 올 확률을 계산한다.

문장을 보면 "나는", "아침에"라는 단어가 이미 나왔다.

"밥을"이라는 단어가 나올 확률을 조건부 확률을 사용하여 계산한다.

### 카운트 기반 접근

문장이나 단어 시퀀스가 얼마나 자주 나타나는지를 '세어서' 그 정보를 바탕으로 문장의 확률을 계산하는 방식


---


아래는 수식은 수학적인 감각이 있어야 하니까 그냥 그림 보듯 보고 넘어가도 된다..

안녕하세요 저는 딥러닝을 공부하고'가 나왔을 때 '있습니다'가 나올 확률을 구해보기.

$P(있습니다|안녕하세요~저는~딥러닝을~공부하고) = {\frac{count(안녕하세요저는딥러닝을공부하고있습니다)}{count(안녕하세요저는딥러닝을공부하고)}}$

단순히 학습한 코퍼스 데이터에서 '안녕하세요 저는 딥러닝을 공부하고'가 100번 등장하고, 그 다음에 '있습니다'가 등장한 경우가 30번이라고 한다면 P(있습니다|안녕하세요저는딥러닝을공부하고)는 30%가 된다.


---



카운트 기반 접근의 기본 원리

단어 빈도수 계산: 카운트 기반 접근에서는 특정 단어나 단어의 조합이 얼마나 자주 등장하는지를 세어본다.

"나는 아침에 밥을 먹는다"라는 문장에서 "아침에 밥을"이라는 단어 조합이 얼마나 자주 나타나는지를 계산

---



카운트 기반 접근은 한계점이 있다.

카운트 기반 조건부 확률을 구하는 공식에서 분모에 해당하는 카운트 수가 0이면 확률이 정의될 수 가 없다.

말이 어렵지?

쉽게 말해, 학습 데이터를 통해 카운트가 한 번도 되지 않거나 아주 적게되는 경우 언어를 정확히 모델링하지 못하는 문제가 생긴다.

이러한 문제를 완화하기 위해 'n-gram'이나 스무딩 백오프와 같은 여러가지 일반화(Generalization) 기법이 존재한다.

하지만 근본적 해결책은 되지 못하기에, 통계적 언어 모델에서 인공 신경망 언어 모델로 트렌트가 바뀌었다.

이런걸 우린 전문적인 말로 '희소 문제' 라고 칭하기로 했다.

'희소 문제' 와 'n-gram' 잠깐 찍먹 수준으로 설명하겠다.

### 희소 문제 개념



데이터나 자원이 한정되어 있어서, 원하는 모든 정보나 결과를 얻기 어려운 상황을 말한다.

드물게 사용되는 단어나 특정한 문장 구조는 데이터 안에서 충분히 나타나지 않을 수 있지 않겠는가?

이런 단어나 문장 구조에 대한 확률을 계산하기 어렵다.

즉, 한정된 데이터로만 학습된 모델은 새로운 데이터나 상황에 잘 적응을 못할 수 있다.

희소 표현의 문제는 학습하다보면 여러번 설명 되니까 너무 어렵게 생각하지 말자.

### N-gram 맛보기

N-gram Language Model은 문장에서 연속적으로 등장하는 N개의 단어를 하나의 단위로 보는 방법

이 단위들의 등장 빈도를 기반으로 문장의 확률을 계산한다.



---





3-gram 모델에서는 문장을 3개의 연속된 단어 조합으로 나누어 각 조합의 빈도를 계산.

통상적으로 2-3개의 단어 까지만 한 조합으로 본다.

한개의 단어의 경우 유니그램

두개의 단어의 경우 바이그램

세개의 단어의 경우 트리그램

이라고 한다.

어렵지????  글만 보고 이해하면 당신은 지니어스!~

아래에 각각 예시를 보여주겠다.

---

유니그램(Unigram)

단어 각각 하나 하나의 등장 확률을 계산할 때 사용.

예시: "나는 학교에 간다"

유니그램은 ['나는', '학교에', '간다'].

나는 : 0.33333..

학교에 : 0.33333..

간다 : 0.3333..

$P(w_1,w_2,\cdots,w_m) = \Pi_{i=1}^m P(w_i)$

수식은 나도 모르겠다..

---



바이그램(Bigram)

텍스트에서 연속적으로 나온 두 개의 단어가 하나의 조합으로 봄.

인접한 두 단어의 등장 확률을 계산할 때 사용.

예시 문장: "나는 학교에 간다"

['나는 학교에', '학교에 간다']

$P(w_1,w_2,\cdots,w_m) = P(w_1) \Pi_{i=2}^m P(w_i {\mid} w_{i-1})$

---






트리그램(Trigram)

텍스트에서 연속적으로 세 개의 단어로 이루어진 조합.

인접한 세 단어의 등장 확률을 계산할 때 사용.

예시 문장 : "나는 오늘 학교에 갔다가 친구를 만났어."

['나는 오늘 학교에', '오늘 학교에 갔다가', '학교에 갔다가 친구를', '갔다가 친구를 만났어']

---



N-gram 수식

$P(w_1,w_2,\cdots,w_m) = P(w_1) \Pi_{i=n}^m P(w_i {\mid} w_{i-1}, \cdots,w_{i-n})$


---




### N-gram 한계점




앞서 말했듯 N-gram은 통계적 접근 기반의 언어 모델이라 희소 문제를 갖고있다.

이 외에도 문맥의 한계, 성능 문제가 있다.

---




n-gram 희소 문제

특정 단어의 조합이 데이터에 충분히 등장하지 않으면, 그 조합의 확률을 제대로 계산할 수 없다.

예를 들어, "나는 아침에 산책을 좋아해"라는 문장에서 "아침에 산책을"이라는 데이터가 없다면, 이 문장의 확률을 정확히 예측하기 어렵다.

---



n-gram 문맥 한계

정해진 N개의 단어만을 고려하기 때문에, 그 이전의 문맥은 고려하지 않는다.

이는 문장이 길어질수록 문맥을 제대로 반영하지 못하는 문제로 이어질 수 있다.


---



n-gram 성능 문제

계산해야 할 확률의 수가 많아질수록, 즉 N의 값이 커질수록 계산량이 기하급수적으로 늘어난다.

이런 극한의 계산량은 모델의 성능을 저하시키는 요인 중 하나이다.

---



N-gram 모델은 언어 모델링의 기초적인 방법 중 하나지만, 위와 같은 한계 때문에 현대의 NLP 연구에서는 더 발전된 모델들이 많이 사용되고 있다.

그러니 아! N-gram 이라는 개념과 방식이 있구나~ 정도만 생각하고 넘어가도록 하자.

원래 조건부확률 계산이라는 어려운 수학적 개념도 있지만 굳이 포함하지 않았다.

---



조건부확률 계산이 궁금하면 수식과 함께 대략적인 설명을 밑에 적긴하겠다.

참고로 난 이해 못하겠다. 그리고 몰라도 큰 문제는 없다.

단어 빈도를 추정하면 각각의 조건부 확률은 기본적으로 다음과 같이 추정을 할 수 있다.

$~~~~~~~~~~~~~~P(w {\mid} w_c){\dfrac {C(w_c,w)}{C(w_c)}}$

위 식에서 $C(w_c,w)$은 말뭉치에서 $(w_c,w)라는 바이그램이 나타내는 횟수 $C(w_c)$은 전체 말뭉치에서 $(W_c) 라는 유니그램(단어)이 나타는 횟수

## 언어모델 평가 개념 정리

모델을 학습하다면 이 모델이 좋은지? 궁금할 것이다.

그렇다면 모델을 평가해보는 것이다.

---


### 분기계수 (Branching Factor)





'언어 평가'를 알기에 앞서 '분기 계수'를 알아야 한다.

---



 어떤 노드(상황)에서 다음 단계로 넘어갈 때 고려할 수 있는 선택지의 수를 의미.

 예를 들어, 어떤 상황에서 3가지의 선택지가 있다면, 분기 계수는 3이 된다.

 이 개념은 나중에 '트리' '분류' 같은 학습에 활용 되는 개념이니까 미리 가볍게 알고가도 좋다.

---



### 펄플렉서티 (Perplexity, PPL)


외부 평가 (extrinsic evalution)

예를 들어 모델 A, B가 있을 때 어떻게 두 모델의 성능을 비교할 수 있을까?

모델의 오타 교정, 기계 번역 등의 평가에 투입해 볼 수 있겠다.

그리고 두 모델의 성능을 누가 더 잘했는지 비교하면 되겠다.

이러한 평가 방식은 외부 평가라고 한다.

---







'외부 평가'보다 부정확 할 수 있지만 테스트 데이터에 대해 빠르게 식으로 계산한 간단한 평가 방법이 있다.

---

펄플렉서티 (Perplexity)



모델 내에서 자신의 성능을 수치화하여 결과를 내놓는 '내부 평가'에 해당하는 '펄플렉서티'이다.

펄플렉서티는 쉽게 말해 내부 평가 지표 이다.

줄여서 PPL 이라 한다.

---



'펄플렉서티'는 언어 모델 성능 측정하는 지표 중 하나이다.

언어 모델이 얼마나 이상하게 예측하는지 숫자로 나타낸다.

즉, 숫자가 낮으면 좋은 예측을 하는 것.

### 국소 표현(이산 표현) (Local Representation)





각 단어들에 고유값에 짝을 짓거나 벡터로 표현하는 방식.

이 방식의 단점은 각 단어 간의 관계나 유사성을 반영하지 못한다는 점이 있다.

종류 : One-Hot Vector, N-Gram, Count base(Bag of Words) 등

그냥 있다 정도만 생각하고 넘어가도 된다.

뒤에 설명할 것이다.

국소 표현의 특징

독립적인 표현: 각 단어는 고유한 벡터로 표현되며, 다른 단어와의 관계를 반영하지 않는다.

간단하고 명확: 단어의 의미나 문맥을 고려하지 않기 때문에, 구현이 간단하고 이해하기 쉽다.

#### Bag of Word (Bow)

장점 : 쉽고 빠른 구축, 예상보다 문서 특징을 잘 나타내어 여러 분야에서 활용도가 높음.

```python
import konlpy
from konlpy.tag import Okt
```

```python
# stopword list
stopword_list = ['을', '은', '는', '이', '가', '를', '있', '하', '것', '들',' 그', '되', '수', '보', '않', '없', ',', '.', "게", '의', '내', '에서', '에']
```

```python
# Okt instance 생성
okt = Okt()
```

```python
# 형태소 단위로 분리 후 stopword 제거하는 함수
def Tokenize(word: str) -> list:
  temp = okt.morphs(word)
  return [k for k in temp if k not in stopword_list]
```

```python
# make Bag of Words
def make_BOW(document: list) -> dict: #주어진 문서에서 고유 단어들의 집합을 생성하고, 각 단어에 고유한 인덱스를 할당.
  words = set(document) # 문서에서 중복을 제거한 고유 단어들의 집합을 생성.
  BOW = dict() #각 단어에 인덱스를 할당하기 위한 빈 딕셔너리를 초기화.
  for word in words: #고유 단어 집합을 순회하며, 각 단어에 고유한 인덱스를 할당.
    if word not in BOW:
      BOW[word] = len(BOW)
  return BOW
```

```python
# 문장을 토큰화해서 리스트에 담은걸 문장1, 문장2, 문장3 에 할당
sentence1 = Tokenize("나는 매일 아침 달리기를 한다")
sentence2 = Tokenize("나는 매일 저녁 헬스장에 간다")
sentence3 = Tokenize("날씨가 많이 추워졌다")
BoW_sentence = sentence1 + sentence2 + sentence3
bows = make_BOW(BoW_sentence)
```

```python
print(bows) # make_BOW 동작 확인
```

```python
# make count vector
def make_cntvec(document: list, BOW: dict) -> list: #주어진 문서에 대해 BoW 모델을 기반으로 각 단어의 출현 빈도를 기록한 벡터를 생성
  count_vector = [0] * len(BOW) #BoW 모델에 있는 각 단어에 대응하는 출현 빈도를 기록할 벡터를 0으로 초기화
  for idx, word in enumerate(BOW.keys()): #BoW 모델의 각 단어에 대해, 해당 단어가 문서 내에서 몇 번 등장하는지 계산하여 count_vector에 기록
    count_vector[idx] = document.count(word)
  return count_vector
```

```python
# 주어진 문장에서 bow에 해당하는 단어 몇번 등장하는지 확인하여 횟수를 측정해 변수에 할당한다.
sent1_cntvec = make_cntvec(sentence1, bows)
sent2_cntvec = make_cntvec(sentence2, bows)
sent3_cntvec = make_cntvec(sentence3, bows)
```

```python
print(list(bows.keys())) #['한다', '달리기', '저녁', '나', '추워졌다', '날씨', '헬스장', '많이', '매일', '아침', '간다']
print(sent1_cntvec) #한다1 달리기1 나1 매일1 아침1
print(sent2_cntvec) #저녁1 나1 저녁1 헬스장1 매일1 간다1
print(sent3_cntvec) #추워졌다1 날씨1 많이1
```

###### CountVectorizer 초기화 파라미터

지금 CountVectorizer의 초기화 파라미터 전부 알려하면 혼란스러울 것이다.

나중에 배울 DTM과 TF-IDF 부분에서 다시 알아볼테니 걱정마라.

그치만 지금은 바로 예제코드만 봐도 알 수 있는 stop_words (불용어 처리)만 예제 알고 넘어가겠다.

그냥 초기화 파라미터가 다양하게 있구나 정도만 생각하자..

바로 아래 내용은 사이킷런 초기화 파라미터 다양하게 있다 정도만 알리기위해 우선 미래 적어놨을 뿐이다. 지금은 큰 의미는 없다.

스르르륵 넘어가자.

---



- max_df : 전체 문서에 걸쳐서 너무 높은 빈도수를 가지는 단어 치러를 제외하기 위한 파라미터, 너무 높은 빈도수를 가지는 단어는 스톱 워드와 비슷한 문법적인 특성으로 반복적인 단어일 가능성이 높기에 이를 제거하기 위해 사용됨. max_df = 100과 같이 정수 값을 가지면 전체 문서에 걸쳐 100개 이하로 나타나는 단어만 피처로 추출. mdf_df = 0.95와 같이 부동소수점(0.0 ~ 1.0)을 가지면 전체 문서에 걸쳐 빈도수 0~95%까지의 단어만 피처로 추출하고 나머지 상위 5%는 피처로 추출하지 않음.

- min_df : 전체 문서에 걸쳐서 너무 낮은 빈도수를 가지는 단어 치러를 제외하기 위한 파라미터. 수백 ~ 수천 개의 전체 문서에서 특정 단어가 min_df에 설정된 값보다 적은 빈도수를 가진다면 이 단어는 크게 중요하지 않거나 필요없는(garbage)성 단어일 확률이 높습니다. min_df = 2와 같이 정수 값을 가지면 전체 문서에 걸쳐서 2번 이하로 나타나는 단어는 리처로 추출하지 않습니다. min_df = 0.002와 같이 부동소수점 값(0.0 ~ 1.0)을 가지면 전체 문서에 걸쳐서 하위 2%이하의 빈도수를 가지는 단어는 피처로 추출하지 않음.
- max_features:  피처로 추출하는 피처의 개수를 제한하며 정수로 값을 지정. 가령 max_features = 2000으로 지정할 경우 가장 높은 빈도를 가지는 단어 순으로 정렬해 2000개까지만 피처로 추출

- stop_words = 'english'로 지정하면 영어의 스톱워드로 지정된 단어는 추출에서 제외.

- ngram_range : bag of words 모델의 단어 순서를 어느 정도 보강하기 위한 n-gram 범위를 설정. 튜플 형태로(범위 최솟값, 범위 최댓값)을 지정. 예를 들어 (1,1)로 지정하면 토큰화된 단어를 1개씩 피처로 추출. (1,2)로 지정하면 토큰화된 단어를 1개씩(minumum 1), 그리고 순서대로 2개씩 (maximum 2) 묶어서 피처로 추출.

- analyzer : 피처 추출을 수행한 단위로 지정. 디폴트는 'word'. word가 아니라 character의 특정 범위를 피처로 만드는 특정한 경우 등을 적용할 때 사용 됨.

- token_patten : 토큰화를 수행하는 정규 표현식 패턴을 지정. 디폴트 값은 '\b\w\w+\b'로, 공백 또는 개행 문자 등으로 구분된 단어 분리자(\b) 사이의 2문자(문자 또는 숫자, 즉 영숫자) 이상의 단어(word)를 토큰으로 분리. analyzer = 'word'로 설정했을 때만 변경 가능하나 디폴트 값을 변경할 경우는 거의 발생 하지 않음. 어근 추출시 외부 함수를 사용할 경우 해당 외부 함수를 token_pattern의 인자로 사용.

- lower_case : 모든 문자를 소문자로 변경할 것인지를 설정.기본 True

###### CountVectorizer 불용어 처리

CountVectorizer의 다양한 초기화 파라미터 중 stop_words 기능만 맛보도록 하겠다.



```python
# 자체 불용어 사용하기
from sklearn.feature_extraction.text import CountVectorizer
text=["Family is not an important thing. It's everything."]
vect = CountVectorizer(stop_words=["the", "a", "an", "is", "not"])
print(vect.fit_transform(text).toarray())
print(vect.vocabulary_)
```

```python
# 사이킷런에서 제공하는 영어 불용어 사용
from sklearn.feature_extraction.text import CountVectorizer
text=["Family is not an important thing. It's everything."]
vect = CountVectorizer(stop_words="english")
print(vect.fit_transform(text).toarray())
print(vect.vocabulary_)
```

```python
# NLTK 지원하는 영어 불용어 사용
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
text=["Family is not an important thing. It's everything."]
sw = stopwords.words("english")
vect = CountVectorizer(stop_words =sw)
print(vect.fit_transform(text).toarray())
print(vect.vocabulary_)
```

파라미터 초기화 알기전에 미리 알아야할 용어 한가지!! 피처

"피처"는 기계 학습에서 사용되는 용어다.

입력 데이터의 특징이나 속성을 의미한다.

텍스트 데이터를 다룰 때, 각각의 단어나 문장 등이 피처가 될 수 있다.

예를 들어, 카운트벡터라이저를 사용할 때는 텍스트 내의 단어들을 피처로 추출하여 이를 숫자 벡터 형태로 변환하는 작업을 수행한다.

이렇게 변환된 각 단어(피처)는 기계 학습 모델이 이해할 수 있는 형태로 입력되어, 모델이 학습하고 예측하는 데 사용된다.

---

max_df: 너무 자주 등장하는 단어는 제외한다.
예를 들어, max_df=0.95는 전체 문서의 95% 이상에 나타나는 단어는 무시한다는 뜻.🚫


---


min_df: 반대로, 거의 등장하지 않는 단어는 무시. min_df=2라면, 2번 이하로 나타나는 단어는 무시한다. 🙅‍♂️


---


max_features: 추출할 피처의 최대 개수를 정한다. 예를 들어, max_features=2000은 빈도수가 가장 높은 2000개의 단어만 선택한다는 의미. 🔢


---


stop_words: 특정 단어를 무시할 수 있다. stop_words='english'는 미리 정의한 영어의 일반적인 불용어를 제외한다는 뜻. 🛑


---


ngram_range: 단어의 조합을 어떻게 추출할지 결정. (1,2)는 단어 하나와 두 단어 조합 모두를 고려한다는 의미. 📊


---


analyzer: 피처 추출의 단위를 정한다. 보통은 'word'로 설정되어 단어를 기준으로 하지만, 문자(character) 단위로도 설정할 수 있다. 🔍


---


token_pattern: 토큰화할 때 사용할 패턴을 정한다. 설정 안하면 기본적으로는 두 글자 이상의 단어를 토큰으로 분리한다. 🧩


---


lower_case: 모든 문자를 소문자로 바꿀지 결정. 보통은 True로 설정되어 있다. 🔡

#### CountVectorizer 사용해보기

```python
from sklearn.feature_extraction.text import CountVectorizer
text_data1=['나는 내일 공부 해야지',
            '내일 날씨는 좋을까',
            '내일 어디를 가지',
            '공부 집에서 해야지']
count_vectorizer1=CountVectorizer() #카운트벡터라이즈 호출
count_vectorizer1.fit(text_data1) #fit 데이터 내의 모든 단어를 스캔하여 각각의 단어에 고유한 정수 인덱스를 할당
print(count_vectorizer1.vocabulary_) # 단어마다 할당된 인덱스 확인 vocabulary_
```

토크나이징은 이미 되어있다고 가정합니다.

```python
import numpy as np

class myCountVectorizer():

    def __init__(self):
        self.vocabulary_ = {}

    def fit(self, docs):
        self.vocabulary_ = {}
        for words in docs:
            for word in words:
                self.vocabulary_[word] = self.vocabulary_.get(word, len(self.vocabulary_))
        return self

    def transform(self, docs):
        m = np.zeros((len(docs), len(self.vocabulary_)))
        for i, words in enumerate(docs):
            for word in words:
                j = self.vocabulary_.get(word)
                if j is not None:
                    m[i, j] += 1
        return m

    def fit_transform(self, docs):
        self.fit(docs)
        return self.transform(docs)
```

fit 단계에서 하는 일은 단어 사전을 만드는 일입니다. 단어마다 인덱스를 부여해서, 이후에 어떤 단어가 행렬의 몇 번 행을 나타내는지 알고자 할 때 사용합니다. 이 단계를 건너 뛰고 이미 만들어져 있는 단어 사전을 이용할 수도 있습니다.

위 코드에서는 딕셔너리의 get 메서드를 이용했습니다. get(k, d)은 딕셔너리에 키가 있으면 해당하는 값을 반환하고, 아니면 d를 반환합니다.

transform 단계에서는 본격적으로 문서-단어 행렬을 만듭니다. 사용법은 CountVectorizer와 비슷합니다.

### DTM의 한계





희소 표현(Sparse representation)

원-핫 벡터는 단어 집합의 크기가 벡터의 차원이 되고 **대부분의 값이 0**이 된다는 특징이 있었다.

이 특징은 **공간적 낭비와 계산 리소스를 증가**시킬 수 있다는 점에서 원-핫 벡터의 단점이었다.

그런 의미에서 DTM도 마찬가지다.


각 문서 벡터의 차원은 원-핫 벡터와 마찬가지로 전체 단어 집합의 크기를 가진다.

만약 가지고 있는 전체 코퍼스가 방대한 데이터라면 문서 벡터의 차원은 수백만의 차원을 가질 수도 있다.


또한 많은 문서 벡터가 대부분의 값이 0을 가질 수도 있다.

당장 위에서 예로 들었던 문서 단어 행렬의 모든 행이 0이 아닌 값보다 0의 값이 더 많은 것을 볼 수 있다.

원-핫 벡터나 DTM과 같은 대부분의 값이 0인 표현을 희소 벡터(sparse vector) 또는 희소 행렬(sparse matrix)라고 부른다.


**희소 벡터는 많은 양의 저장 공간과 계산을 위한 리소스를 필요로 한다.**

이러한 이유로 전처리를 통해 단어 집합의 크기를 줄이는 일은 BoW 표현을 사용하는 모델에서 중요할 수 있다.


텍스트 전처리 방법을 사용하여 구두점, 빈도수가 낮은 단어, 불용어를 제거하고, 어간이나 표제어 추출을 통해 단어를 정규화하여 단어 집합의 크기를 줄일 수 있다.

그치만 아무리 전처리 작업해도 한계는 여전히 존재한다.

## TF-IDF (Term Frequency-Inverse Document Frequency) 개념





DTM 한계에 대한 대안책으로 나온게 TF-IDF 이다.

---



각 문서에서 해당 단어가 나타나는 횟수를 단어에 count를 부여하는것을 카운트 벡터화라고 한다.

카운트 벡터화에서는 카운트 값이 높을수록 중요한 단어라고 인식하여 문제가 됨.

영어에 대해 DTM을 만들었을 때, (불용어 처리를 안했다고 가정하면) 'the'는 어떤 문서이든 자주 등장할 수 밖에 없다. *The 뿐만이 아니라 비슷한 경우가 많다.

아무튼 그래서 유사한 문서인지 비교할 때, 단순 the가 빈도수가 높다고 유사한 문서라고 판단해선 안 된다.

불용어와 중요 단어에 대해 가중치를 줄 수 있는 방법은 없을까?

그래서 나온게 TF-IDF 라는 개념이다.
